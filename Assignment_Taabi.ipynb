## Outlier detection using box plot

```python
import math

# Calculate the number of rows needed for the subplots
num_cols = len(continuous_columns)
num_rows = math.ceil(num_cols / 3)  # 3 columns per row

plt.figure(figsize=(15, num_rows * 5))  # Adjust height based on the number of rows
for i, col in enumerate(continuous_columns, 1):
    plt.subplot(num_rows, 3, i)
    sns.boxplot(x=df[col])
    plt.title(col)
plt.tight_layout()
plt.show()
```

```python
df.coolant.describe()
```

```python
Q1 = 83  # 25th percentile
Q3 = 89  # 75th percentile
IQR = Q3 - Q1

# Define the lower and upper bounds for outliers
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Check if there are outliers
if (30 < lower_bound) or (98 > upper_bound):
    print("Outliers are present in the 'coolant' column.")
else:
    print("No outliers are present in the 'coolant' column.")
```

```python
# Filter out rows containing outliers
coolant_no_outliers = df[(df['coolant'] >= lower_bound) & (df['coolant'] <= upper_bound)]
```

```python
# Filter out rows containing outliers
df= df[ (df.coolant >= lower_bound) & (df.coolant <= upper_bound)] 
```
